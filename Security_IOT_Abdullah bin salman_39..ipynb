```python
from cryptography.hazmat.primitives.ciphers import Cipher, algorithms, modes
from cryptography.hazmat.backends import default_backend
import os
import base64
import json
import time
from datetime import datetime

print("=== PART I: IoT Data Encryption Simulation ===")

class IoTSensor:
    def __init__(self, sensor_id):
        self.sensor_id = sensor_id
        self.encryption_key = os.urandom(16)  # 128-bit key

    def generate_sensor_data(self):
        import random
        temperature = round(random.uniform(20.0, 35.0), 2)
        humidity = round(random.uniform(30.0, 80.0), 2)
        return {
            'temperature': temperature,
            'humidity': humidity,
            'sensor_id': self.sensor_id,
            'timestamp': datetime.now().isoformat()
        }

    def encrypt_data(self, data):
        # Convert data to JSON string
        data_str = json.dumps(data)
        data_bytes = data_str.encode('utf-8')

        # Generate random IV
        iv = os.urandom(16)

        # Pad data to be multiple of 16 bytes
        pad_length = 16 - (len(data_bytes) % 16)
        data_bytes += bytes([pad_length] * pad_length)

        # Create cipher and encrypt
        cipher = Cipher(algorithms.AES(self.encryption_key), modes.CBC(iv), backend=default_backend())
        encryptor = cipher.encryptor()
        encrypted_data = encryptor.update(data_bytes) + encryptor.finalize()

        # Return IV + encrypted data
        return base64.b64encode(iv + encrypted_data).decode('utf-8')

    def decrypt_data(self, encrypted_data):
        # Decode from base64
        encrypted_bytes = base64.b64decode(encrypted_data.encode('utf-8'))

        # Extract IV and encrypted data
        iv = encrypted_bytes[:16]
        encrypted_data = encrypted_bytes[16:]

        # Create cipher and decrypt
        cipher = Cipher(algorithms.AES(self.encryption_key), modes.CBC(iv), backend=default_backend())
        decryptor = cipher.decryptor()
        decrypted_bytes = decryptor.update(encrypted_data) + decryptor.finalize()

        # Remove padding
        pad_length = decrypted_bytes[-1]
        decrypted_bytes = decrypted_bytes[:-pad_length]

        # Convert back to dictionary
        return json.loads(decrypted_bytes.decode('utf-8'))

# Ø§Ù„ØªÙ†ÙÙŠØ°
sensor = IoTSensor("sensor_001")
original_data = sensor.generate_sensor_data()
print("ğŸ“Š Original Data:")
print(f"  Temperature: {original_data['temperature']}Â°C")
print(f"  Humidity: {original_data['humidity']}%")
print(f"  Sensor ID: {original_data['sensor_id']}")

encrypted_data = sensor.encrypt_data(original_data)
print(f"\nğŸ”’ Encrypted Data (first 50 chars): {encrypted_data[:50]}...")

decrypted_data = sensor.decrypt_data(encrypted_data)
print("\nğŸ“¥ Decrypted Data:")
print(f"  Temperature: {decrypted_data['temperature']}Â°C")
print(f"  Humidity: {decrypted_data['humidity']}%")

if original_data['temperature'] == decrypted_data['temperature']:
    print("\nâœ… Data integrity verified successfully!")
else:
    print("\nâŒ Data integrity check failed!")
```
